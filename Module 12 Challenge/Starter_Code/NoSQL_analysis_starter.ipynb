### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

#degree_search = 0.01
#latitude =
#longitude =

#query =
#sort =
#limit =

# Print the results
import pprint

# Given latitude and longitude of "Penang Flavours"
new_restaurant_lat = 51.490142
new_restaurant_lon = 0.08384

# Define the degree range for latitude and longitude search
degree_search = 0.01

# Define the query to find establishments with RatingValue of 5, within the degree range
query = {
    "RatingValue": 5,
    "scores.Hygiene": {"$ne": None},
    "geocode.latitude": {"$gte": new_restaurant_lat - degree_search, "$lte": new_restaurant_lat + degree_search},
    "geocode.longitude": {"$gte": new_restaurant_lon - degree_search, "$lte": new_restaurant_lon + degree_search}
}

# Define the sort order to sort by lowest hygiene score
sort = [("scores.Hygiene", 1)]

# Define the limit to retrieve only the top 5 establishments
limit = 5

# Execute the query
top_5_establishments = establishments.find(query).sort(sort).limit(limit)

# Print the results
for idx, establishment in enumerate(top_5_establishments, 1):
    print(f"Establishment {idx}:")
    pprint.pprint(establishment)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest

# Print the number of documents in the result

# Print the first 10 results

# Create the aggregation pipeline
pipeline = [
    {"$match": {"scores.Hygiene": 0}},
    {"$group": {"_id": "$LocalAuthorityName", "count": {"$sum": 1}}},
    {"$sort": {"count": -1}}
]

# Execute the aggregation pipeline
result = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of establishments in each Local Authority area with a hygiene score of 0:")
for doc in result:
    print(f"{doc['_id']}: {doc['count']}")

# Print the first 10 results
print("\nFirst 10 results:")
for doc in result[:10]:
    print(doc)

```

```python
# Convert the result to a Pandas DataFrame

# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(result)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {len(df)}")

# Display the first 10 rows of the DataFrame
print("\nFirst 10 rows of the DataFrame:")
print(df.head(10))
```
